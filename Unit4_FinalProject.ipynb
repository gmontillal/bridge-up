## Finding absolute magnitude

You will need to convert "G" to absolute magnitude for your CMD.

Look in the dataset for distance. Using G and distance, you can calculate absolute magnitude using the following equation:

absolute magnitude = G - 5 * log(distance) + 5

```python
# You will need to import numpy
import numpy as np
```

```python
# You can use this function in your code to find the log of distance:
log_d = np.log10(df["distance"])

df["absolute magnitude"] = df['G'] - 5 * log_d + 5
df.head()
# np.log10()

gC = df['G']
jC = df['J']
abs_mag = df['G'] - 5 * log_d + 5
```

### Your first task is to create a Color Magnitude Diagram for your star cluster
You can use as many of the cells below as you need. You can add new cells with the "+" button.

```python
import matplotlib.colors as mpl
%matplotlib inline 
gj = gC - jC
len(gj)
```

```python
cm = plt.cm.get_cmap("RdYlBu_r")
plt.style.use('dark_background')
plt.scatter(gC - jC, abs_mag, c = gC - jC, s = 35, cmap = cm, norm=mpl.LogNorm()) 
plt.xlabel('Color (G-J)')
plt.ylabel('Absolute Magnitude (G)', color = "violet")
plt.title('Color Magnitude Diagram')
plt.gca().invert_yaxis()
#cm is the color map
#style is for the black background
#"gC - jC" is the x axis and "abs_mag" is the yaxis/s =size of the dots/cmap = color map's colors/"norm=mpl.LogNorm()
# gives us the library for the colors 
#c is being guided by the x axis so it will color code it depending on the x axis
# g= gaia/ optical range 
#j = infrared [they are two different wavelenghts]
#plt.gca inverts the yaxis
```
