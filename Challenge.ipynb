Import

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Open file

```python
df = pd.read_csv("SeaTemp_SeaLevel_anomaly.csv")
```

Column names

```python
col_year = df["Year"]
anom_col = df["Sea temperature anomaly (C)"]
temp_col = df["Change in mean sea level (mm)"]
```

Graph


```python
plt.plot(col_year, anom_col, color = "purple")
plt.legend()
plt.ylabel("Anomalies Temp (celsius)")
plt.title("Sea Temp vs. Mean Sea Temp")
plt.xlabel("Years")
plt.twinx()
plt.plot(col_year, temp_col, color = "red")
plt.ylabel("Sea Temp (celsius)")
plt.legend(loc = "lower center")
plt.show()
```
