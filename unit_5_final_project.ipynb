```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('bird_dino_data.csv')
```

```python
df.head()
```

```python
# add a new column to the df
df['Brain Volume (cm3)'] = df['Olfactory bulbs (cm3)'] + df['Cerebrum (cm3)'] + df['Optic Lobes (cm3)'] + df['Cerebellum (cm3)']
df.head()
```

```python
#new column whole brain to body ratio
#cm3 = g

df["brain to body R"] = df['Brain Volume (cm3)']/ df['Body Mass (kg)']/1000
df.head()
```

```python
# column with cerebrum to whole brain ratio
df["Cerebrum to brain R"] = df["Cerebrum (cm3)"]/ df['Brain Volume (cm3)']
df.head()
```

```python
# calc the avg of birds and dinos
B = df.loc[df['Bird vs Dino'] == "B"]
D = df.loc[df['Bird vs Dino'] == "D"]
B.head()
```

```python
D.head(9)
```

```python
btbB = B['brain to body R'].mean()
```

```python
btbD = D['brain to body R'].mean()
```

```python
#function
def dinosaur(dtype, femurl):
    if dtype == "ornithischian":
        mass1 = 10**(3.0587*np.log10(femurl)-2.7042)/1000
    elif dtype == "sauropod":
        mass1 = 10**(2.3459*np.log10(femurl)-0.2935)/1000
    elif dtype == "theropod":
        mass1=10**(3.1854*np.log10(femurl)-3.184)/1000
    return mass1
```

```python
#use in cm
bodyM = dinosaur("theropod", 21)
print(bodyM)
```

```python
body_m_g = bodyM * 1000
```

```python
 def compare(FL, cerebrumW, brainW):
    mass = dinosaur("theropod", FL)
    brain2body = brainW/mass*1000
    cerebrum2brain = cerebrumW/brainW
    if abs(myB_to_body-btbB)> abs(myB_to_body-btbD):
        b2b_ans = "Dino"
    else:
        b2b_ans = "Bird"
    
    if abs(myB_to_body-btbB)< abs(myB_to_body-btbD):
        c2b_ans = "Dino"
    else:
        c2b_ans = "Bird"
    return (b2b_ans, c2b_ans) 
```
